```python
import pandas as pd
import pycaret as pc
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/olympics.csv')
df['medal']=df['medal'].fillna(0)
df_orig = df
df_orig
```

```python
df = df_orig
df.query('medal!=0', inplace=True)
df.query('season=="Summer"', inplace=True)
df.query('sport=="Athletics"', inplace=True)
df['height'] = df["height"].apply(lambda x: x * 0.0328)
df = df.dropna(subset=['height'])
df = df[['sex','height','year','sport']]
```

```python
df_men = df_orig.query("sex=='M'").drop(columns=['sex'])
df_men = df_men[['event','height','year']] #df_men.query('event.str.contains("100")', engine = 'python')[['event','height','year']]
df_men_group = df_men.groupby(['event','year']).max()
df_men_group = df_men_group.reset_index()
fig = px.line(df_men_group,x='year',y='height', color='event')
fig.show()
```

```python
df_women = df_orig.query("sex=='F'").drop(columns=['sex'])
df_women = df_women[['event','height','year']] #df_women.query('event.str.contains("100")', engine = 'python')[['event','height','year']]
df_women_group = df_women.groupby(['event','year']).max()
df_women_group = df_women_group.reset_index()
fig = px.line(df_women_group,x='year',y='height', color='event')
fig.show()
```

```python
df_bell = df_orig
df['height']=df['height'].fillna(0)
mean_val = np.mean((df_bell['height']))
std_val = np.std((df_bell['height']))
median_val = np.median((df_bell['height']))
print(str.format("mean:{},std:{},median:{}",mean_val, std_val, median_val))
fig = ff.create_distplot([df['height']], ['distplot'], curve_type='normal', histnorm='probability density')
fig.add_shape(type='line',x0=mean_val,x1=mean_val,y0=0,y1=0.7,xref='x',yref='y',line=dict(color='blue', dash='dash'))
fig.add_shape(type='line',x0=median_val,x1=median_val,y0=0,y1=0.7,xref='x',yref='y',line=dict(color='green', dash='dash'))
fig.add_shape(type='line',x0=mean_val+std_val,x1=mean_val+std_val,y0=0,y1=0.7,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.add_shape(type='line',x0=mean_val+(std_val + std_val),x1=mean_val+(std_val + std_val),y0=0,y1=0.7,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.add_shape(type='line',x0=mean_val-std_val,x1=mean_val-std_val,y0=0,y1=0.7,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.show()
```
