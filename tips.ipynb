```python
import pandas as pd
import pycaret as pc
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
```

```python
df = px.data.tips()
df
```

```python
mean_val = np.mean((df['tip']))
std_val = np.std((df['tip']))
mins_std_val = np.std((df['tip']))*-1
fig = ff.create_distplot([df['tip']], ['distplot'], curve_type='normal')
fig.add_shape(type='line',x0=mean_val,x1=mean_val,y0=0,y1=0.4,xref='x',yref='y',line=dict(color='blue', dash='dash'))
fig.add_shape(type='line',x0=mean_val+std_val,x1=mean_val+std_val,y0=0,y1=0.4,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.add_shape(type='line',x0=mean_val+(std_val + std_val),x1=mean_val+(std_val + std_val),y0=0,y1=0.4,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.add_shape(type='line',x0=mean_val-std_val,x1=mean_val-std_val,y0=0,y1=0.4,xref='x',yref='y',line=dict(color='red', dash='dash'))
fig.show()
```
